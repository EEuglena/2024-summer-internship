```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
size = [1000, 5000, 10000, 15000, 25000]
error = [616.41, 417.76, 139.71, 121.24, 34.61]
```

```python
plt.plot((size), np.log(error))
plt.scatter((size), np.log(error))
plt.xlabel("Size of dataset")
plt.ylabel("log(RMSE)")
plt.show()
```

```python
plt.plot((size), (error))
plt.scatter((size), (error))
plt.xlabel("Size of dataset")
plt.ylabel("RMSE")
plt.show()
```

```python
r2_score(np.log(error), size[::-1])
```
